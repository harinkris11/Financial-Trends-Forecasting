**SENTIMENT ANALYSIS FUNCTION**

```python
def sentimentLabelandScore(classifier_model, df, columnName):
    # df = df[df[columnName].notna()]
    classifier = pipeline("sentiment-analysis", model=classifier_model)
    df[[columnName+' sentiment_label', columnName+' sentiment_score']] = df[columnName].apply(lambda text: pd.Series(classifier(text)[0]))
    return df
```

```python
def loadCSVFromGDrive(path,rowsLimit=None):
  if rowsLimit is None:
    return pd.read_csv(path)
  else:
    return pd.read_csv(path).sample(rowsLimit)
```
